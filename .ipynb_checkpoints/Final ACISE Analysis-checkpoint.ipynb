```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Configure visualization settings
plt.style.use('seaborn-v0_8') if 'seaborn-v0_8' in plt.style.available else plt.style.use('ggplot')
sns.set_style("whitegrid")
sns.set_palette("husl")

def analyze_data(df):
    """Perform comprehensive analysis"""
    if df is None:
        print("No data to analyze")
        return
    
    # Perform statistical tests first to get p-values
    anova_result = perform_anova_test(df, silent=True)
    chi_result = perform_chi_squared_test(df, silent=True)
    
    print("\n=== ANALYSIS RESULTS ===")
    
    # Basic statistics
    print(f"üìä Total records: {len(df):,}")
    print(f"üåç Unique regions: {df['PostalRegion'].nunique()}")
    print(f"üí∞ Avg profit margin: ${df['ProfitMargin'].mean():,.2f}")
    
    # Handle loss ratio calculation
    valid_loss = df[df['TotalPremium'] > 0]
    loss_ratio = valid_loss['LossRatio'].mean() if len(valid_loss) > 0 else np.nan
    if not np.isnan(loss_ratio):
        print(f"üìâ Avg loss ratio: {loss_ratio:.2%}")
    else:
        print("üìâ No valid policies with positive premium for loss ratio calculation")
    
    # Top/bottom performing regions
    region_stats = df.groupby('PostalRegion').agg({
        'ProfitMargin': 'mean',
        'LossRatio': 'mean',
        'TotalPremium': 'count'
    }).rename(columns={'TotalPremium': 'PolicyCount'})
    
    print("\nüèÜ Top 5 Performing Regions:")
    display(region_stats.nlargest(5, 'ProfitMargin'))
    
    print("\n‚ö†Ô∏è Bottom 5 Performing Regions:")
    display(region_stats.nsmallest(5, 'ProfitMargin'))
    
    # Additional insights
    print("\nüîç Additional Insights:")
    print(f"Policies with zero premium: {(df['TotalPremium'] == 0).sum():,}")
    print(f"Policies with zero claims: {(df['TotalClaims'] == 0).sum():,}")
    print(f"Average sum insured: ${df['SumInsured'].mean():,.2f}")

    # Visualizations
    visualize_data(df)
    
    # Comprehensive summary
    print_comprehensive_summary(df, anova_result, chi_result, loss_ratio)
    
    # Business recommendations
    print_business_recommendations(anova_result['p_value'], chi_result['p_value'])
    
    # Final visualization
    plot_final_visualization(df)

def print_comprehensive_summary(df, anova_result, chi_result, loss_ratio):
    """Print formatted summary of all key findings"""
    province_stats = df.groupby('Province')['ProfitMargin'].mean()
    
    print("""
=== ANALYSIS RESULTS SUMMARY ===

1. Key Metrics:
   - Average Profit Margin: ${avg_margin:,.2f}
   - Average Loss Ratio: {loss_ratio:.2%}
   - Median Profit Margin: ${median_margin:,.2f}

2. Geographic Analysis:
   - Highest Margin Province: {top_province} (${top_prov_margin:,.2f})
   - Lowest Margin Province: {bottom_province} (${bottom_prov_margin:,.2f})
   - Regional Variation: {regional_sig} (ANOVA p-value: {p_val_anova:.4f})

3. Statistical Findings:
   - Region-Profit Association: {chi_result} (Chi-sq p-value: {p_val_chi:.4f})
   - Effect Size (Cramer's V): {effect_size:.4f}

4. Data Quality:
   - Analyzed {n_policies:,} policies
   - Covered {n_regions} postal regions
   - {n_provinces} provinces represented
""".format(
        avg_margin=df['ProfitMargin'].mean(),
        loss_ratio=loss_ratio if not np.isnan(loss_ratio) else 0,
        median_margin=df['ProfitMargin'].median(),
        top_province=province_stats.idxmax(),
        top_prov_margin=province_stats.max(),
        bottom_province=province_stats.idxmin(),
        bottom_prov_margin=province_stats.min(),
        regional_sig="Significant" if anova_result['p_value'] < 0.05 else "Not significant",
        p_val_anova=anova_result['p_value'],
        chi_result="Significant" if chi_result['p_value'] < 0.05 else "Not significant",
        p_val_chi=chi_result['p_value'],
        effect_size=chi_result['effect_size'],
        n_policies=len(df),
        n_regions=df['PostalRegion'].nunique(),
        n_provinces=df['Province'].nunique()
    ))

def print_business_recommendations(anova_p, chi_p):
    """Generate actionable business recommendations"""
    if anova_p < 0.05 or chi_p < 0.05:
        print("""
BUSINESS RECOMMENDATIONS FOR SIGNIFICANT FINDINGS
------------------------------------------------
1. GEOGRAPHIC STRATEGY:
   - Implement regional pricing tiers based on profitability
   - Develop targeted marketing for high-margin areas
   - Allocate fraud prevention resources strategically

2. PRODUCT DEVELOPMENT:
   - Analyze coverage preferences by region
   - Develop region-specific product bundles
   - Adjust commissions to incentivize profitable regions

3. OPERATIONAL EXCELLENCE:
   - Benchmark claims processing by region
   - Share best practices across regions
   - Develop regional performance dashboards
""")
    else:
        print("""
BUSINESS RECOMMENDATIONS FOR NON-SIGNIFICANT FINDINGS
-----------------------------------------------------
1. CUSTOMER-CENTRIC APPROACH:
   - Focus on customer segments rather than geography
   - Develop personalized pricing models
   - Analyze behavioral patterns across demographics

2. PRODUCT INNOVATION:
   - Differentiate through coverage enhancements
   - Develop value-added services
   - Implement dynamic pricing models

3. OPERATIONAL EFFICIENCY:
   - Standardize processes across all regions
   - Implement centralized fraud detection
   - Optimize claims handling workflow
""")

def plot_final_visualization(df):
    """Generate final insightful visualization"""
    plt.figure(figsize=(12, 8))
    sample_df = df.sample(min(10000, len(df)))  # Sample for large datasets
    sns.scatterplot(
        x='SumInsured', 
        y='ProfitMargin', 
        hue='PostalRegion',
        size='TotalPremium',  # Add size dimension
        data=sample_df, 
        alpha=0.6, 
        palette='viridis',
        sizes=(20, 200)  # Range of point sizes
    )
    plt.title('Profitability Analysis: Margin vs Sum Insured by Region', pad=20)
    plt.xlabel('Sum Insured ($)', labelpad=10)
    plt.ylabel('Profit Margin ($)', labelpad=10)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Postal Region')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def perform_anova_test(df, silent=False):
    """Perform ANOVA test and return results"""
    if not silent:
        print("\n=== ANOVA TEST: PROFIT MARGIN ACROSS REGIONS ===")
    
    # Select top regions with sufficient data
    top_regions = df['PostalRegion'].value_counts().nlargest(5).index
    filtered_df = df[df['PostalRegion'].isin(top_regions)]
    
    # Prepare data for ANOVA
    groups = [filtered_df[filtered_df['PostalRegion'] == region]['ProfitMargin'] 
              for region in top_regions]
    
    # Perform test
    f_stat, p_value = stats.f_oneway(*groups)
    
    result = {
        'f_stat': f_stat,
        'p_value': p_value,
        'is_significant': p_value < 0.05
    }
    
    if not silent:
        print(f"F-statistic: {f_stat:.4f}, p-value: {p_value:.4f}")
        if p_value < 0.05:
            print("‚úÖ Significant differences exist between regions")
            # Post-hoc analysis would go here
        else:
            print("‚ùå No significant differences between regions")
    
    return result

def perform_chi_squared_test(df, silent=False):
    """Perform Chi-Squared test and return results"""
    if not silent:
        print("\n=== CHI-SQUARED TEST: HIGH-PROFIT POLICIES BY REGION ===")
    
    # Create binary target variable
    median_margin = df['ProfitMargin'].median()
    df['HighProfit'] = df['ProfitMargin'] > median_margin
    
    # Filter regions with sufficient data
    min_policies = 1000
    region_counts = df['PostalRegion'].value_counts()
    valid_regions = region_counts[region_counts >= min_policies].index
    contingency_table = pd.crosstab(
        df[df['PostalRegion'].isin(valid_regions)]['PostalRegion'],
        df[df['PostalRegion'].isin(valid_regions)]['HighProfit']
    )
    
    # Perform test
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    
    # Calculate effect size
    n = contingency_table.sum().sum()
    cramers_v = np.sqrt(chi2 / (n * min(contingency_table.shape) - 1))
    
    result = {
        'chi2': chi2,
        'p_value': p,
        'dof': dof,
        'effect_size': cramers_v,
        'is_significant': p < 0.05
    }
    
    if not silent:
        print(f"Tested {len(valid_regions)} regions with ‚â•{min_policies} policies")
        print(f"œá¬≤({dof}) = {chi2:.2f}, p = {p:.4f}, Cramer's V = {cramers_v:.3f}")
        if p < 0.05:
            print("‚úÖ Significant association between region and profitability")
        else:
            print("‚ùå No significant association found")
    
    return result

# Execute the pipeline
print("=== INSURANCE DATA ANALYSIS ===")
insurance_df = process_insurance_data('insurance_claims.csv')

if insurance_df is not None:
    analyze_data(insurance_df)
else:
    print("Processing failed - no results available")
```
