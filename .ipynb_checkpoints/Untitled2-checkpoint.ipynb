```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_classification
import statsmodels.api as sm
import statsmodels.formula.api as smf

# ==========================================
# 1. DATA LOADING WITH ERROR HANDLING
# ==========================================

def load_data(file_path):
    """Function to safely load the dataset with error handling."""
    try:
        df = pd.read_csv(file_path, sep='|', encoding='ISO-8859-1')
        print("✅ Data loaded successfully")
        return df
    except FileNotFoundError:
        print("❌ Error: File not found. Please check the file path.")
        return None
    except pd.errors.ParserError:
        print("❌ Error: Failed to parse the file. Ensure it's in CSV format.")
        return None
    except Exception as e:
        print(f"❌ Unexpected Error: {str(e)}")
        return None

# Load the dataset
df = load_data('insurance_claims.csv')

# Exit if data wasn't loaded
if df is None:
    exit()

# Check basic info
print("\n=== DATA STRUCTURE ===")
print(df.info())

# Calculate Loss Ratio if needed
if 'TotalPremium' in df.columns and 'TotalClaims' in df.columns:
    df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
else:
    print("❌ Missing required columns for analysis")
    exit()
```

```python
# ======================
# 3. EXPLORATORY DATA ANALYSIS (MEMORY-OPTIMIZED)
# ======================
print("\n=== STEP 3: EXPLORATORY DATA ANALYSIS ===")

def safe_plot(plot_func, *args, **kwargs):
    """Wrapper function to handle memory errors in plotting"""
    try:
        return plot_func(*args, **kwargs)
    except MemoryError:
        print(f"MemoryError in plotting. Trying alternative approach...")
        # Reduce data size by 50% and try again
        if 'data' in kwargs and isinstance(kwargs['data'], pd.DataFrame):
            kwargs['data'] = kwargs['data'].sample(frac=0.5)
        elif len(args) > 0 and isinstance(args[0], pd.Series):
            args = (args[0].sample(frac=0.5),) + args[1:]
        return plot_func(*args, **kwargs)

# 3.1 Univariate Analysis
print("\n--- Univariate Analysis ---")

# Numerical distributions (with memory optimization)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
safe_plot(sns.histplot, df['TotalPremium'], kde=False, color='blue', bins=30)  # Disabled KDE for memory
plt.title('Total Premium Distribution')

plt.subplot(1, 2, 2)
safe_plot(sns.histplot, df['TotalClaims'], kde=False, color='red', bins=30)  # Disabled KDE for memory
plt.title('Total Claims Distribution')
plt.tight_layout()
plt.show()

# Categorical distributions (less memory intensive)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
df['Gender'].value_counts().plot(kind='bar', color='green')
plt.title('Gender Distribution')

plt.subplot(1, 2, 2)
df['Province'].value_counts().plot(kind='bar', color='purple')
plt.title('Province Distribution')
plt.tight_layout()
plt.show()

# Calculate Loss Ratio with error handling
try:
    df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
except Exception as e:
    print(f"Error calculating LossRatio: {str(e)}")
    df['LossRatio'] = np.nan  # Create column with NaN values if calculation fails

# Loss Ratio by Province (optimized version)
plt.figure(figsize=(14, 7))
try:
    province_data = df.groupby('Province')['LossRatio'].mean().reset_index()
    province_plot = sns.barplot(
        x='Province', 
        y='LossRatio', 
        data=province_data,  # Use pre-aggregated data
        palette='viridis',
        estimator=np.mean,
        errorbar=None
    )
    
    plt.title('Loss Ratio by Province (Higher = More Risky)', fontsize=16)
    plt.xticks(rotation=45)
    
    # Add value labels if not too many provinces
    if len(province_data) < 20:
        for p in province_plot.patches:
            province_plot.annotate(
                f"{p.get_height():.2f}", 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points'
            )
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error creating province plot: {str(e)}")

# Correlation Analysis (optimized)
plt.figure(figsize=(10, 8))
try:
    # Select only numerical columns to avoid memory issues
    num_cols = ['TotalPremium', 'TotalClaims', 'CustomValueEstimate']
    corr_matrix = df[num_cols].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Matrix')
    plt.show()
except Exception as e:
    print(f"Error creating correlation matrix: {str(e)}")

# 3.3 Temporal Analysis (optimized)
print("\n--- Temporal Analysis ---")

try:
    # Downsample if too many time points
    if df['TransactionMonth'].nunique() > 24:  # More than 2 years of monthly data
        monthly_data = df.resample('Q', on='TransactionMonth').agg({
            'TotalPremium': 'sum',
            'TotalClaims': 'sum'
        }).reset_index()
        freq_label = 'Quarterly'
    else:
        monthly_data = df.groupby('TransactionMonth').agg({
            'TotalPremium': 'sum',
            'TotalClaims': 'sum'
        }).reset_index()
        freq_label = 'Monthly'

    plt.figure(figsize=(14, 6))
    plt.plot(monthly_data['TransactionMonth'], monthly_data['TotalPremium'], label='Total Premium')
    plt.plot(monthly_data['TransactionMonth'], monthly_data['TotalClaims'], label='Total Claims')
    plt.title(f'{freq_label} Premium vs Claims Trend')
    plt.xlabel('Date')
    plt.ylabel('Amount')
    plt.legend()
    plt.show()
except Exception as e:
    print(f"Error in temporal analysis: {str(e)}")

print("\nEDA completed with memory optimizations")
```

```python
# Compute correlation matrix
def generate_correlation_matrix():
    """Function to safely generate a correlation matrix."""
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    if not numeric_cols.empty:
        corr_matrix = df[numeric_cols].corr()
        plt.figure(figsize=(12, 8))
        sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0)
        plt.title('Correlation Matrix')
        plt.show()
    else:
        print("⚠️ No numeric columns found for correlation analysis")

# Generate correlation matrix
generate_correlation_matrix()
```

```python
def perform_t_test(column_name, group_column, group1, group2):
    """Function to safely perform a T-test for two groups."""
    if column_name in df.columns and group_column in df.columns:
        group1_data = df[df[group_column] == group1][column_name]
        group2_data = df[df[group_column] == group2][column_name]

        if len(group1_data) > 1 and len(group2_data) > 1:
            t_stat, p_value = stats.ttest_ind(group1_data, group2_data, equal_var=True)

            print(f"T-statistic: {t_stat:.4f}")
            print(f"P-value: {p_value:.4f}")

            alpha = 0.05
            if p_value < alpha:
                print("✅ Significant difference detected.")
            else:
                print("❌ No significant difference found.")
        else:
            print(f"⚠️ Insufficient data for T-test in groups '{group1}' and '{group2}'")
    else:
        print(f"⚠️ Missing required columns: '{column_name}' or '{group_column}'")

# Perform T-test
perform_t_test('TotalClaims', 'Gender', 'Male', 'Female')
```

```python
def perform_anova_test(column_name, group_column):
    """Function to safely perform an ANOVA test across multiple categories."""
    if column_name in df.columns and group_column in df.columns:
        unique_groups = df[group_column].unique()
        group_data = [df[df[group_column] == group][column_name] for group in unique_groups]

        if all(len(group) > 1 for group in group_data):
            f_stat, p_value = stats.f_oneway(*group_data)

            print(f"F-statistic: {f_stat:.4f}")
            print(f"P-value: {p_value:.4f}")

            alpha = 0.05
            if p_value < alpha:
                print("✅ Significant differences detected.")
            else:
                print("❌ No significant difference found.")
        else:
            print("⚠️ Insufficient data for ANOVA test.")
    else:
        print(f"⚠️ Missing required columns: '{column_name}' or '{group_column}'")

# Perform ANOVA test
perform_anova_test('TotalClaims', 'Province')
```

```python
def perform_chi_squared_test(column1, column2):
    """Function to safely perform a Chi-Squared test."""
    if column1 in df.columns and column2 in df.columns:
        contingency_table = pd.crosstab(df[column1], df[column2])

        chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

        print(f"Chi-Squared Statistic: {chi2_stat:.4f}")
        print(f"P-Value: {p_val:.4f}")

        alpha = 0.05
        if p_val < alpha:
            print("✅ Significant association detected.")
        else:
            print("❌ No significant association found.")
    else:
        print(f"⚠️ Missing required columns: '{column1}' or '{column2}'")

# Perform Chi-Squared test
perform_chi_squared_test('has_anti_theft_device', 'HadClaim')
```
