## 1. Environment Setup

```python
# Core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Machine Learning
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (mean_absolute_error, 
                           mean_squared_error, 
                           r2_score,
                           make_scorer)

# Models
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import shap

# Configuration
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
np.random.seed(42)
```

## 2. Data Preparation

```python
# Load processed data
df = pd.read_csv('../data/processed/processed_insurance_data.csv')

# Feature Selection
features = [
    'vehicletype', 'vehicle_age', 'suminsured', 
    'province', 'gender', 'make', 'age_group'
]

target = 'totalclaims'

# Filter available features
available_features = [f for f in features if f in df.columns]
X = df[available_features]
y = df[target]

# Train-test split (stratified by claim amount bins)
bins = pd.qcut(y, q=5, labels=False)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=bins, random_state=42
)

print(f"Training samples: {len(X_train):,}")
print(f"Test samples: {len(X_test):,}")
```

## 3. Preprocessing Pipeline

```python
# Numeric vs Categorical features
numeric_features = X_train.select_dtypes(include=['number']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object', 'category']).columns.tolist()

# Pipeline construction
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Example of transformed output
sample_transformed = preprocessor.fit_transform(X_train[:3])
print("Sample transformed data shape:", sample_transformed.shape)
```

## 4. Model Training & Evaluation

```python
# Model Definitions
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(
        n_estimators=150,
        max_depth=10,
        random_state=42
    ),
    "XGBoost": XGBRegressor(
        objective='reg:squarederror',
        n_estimators=200,
        max_depth=6,
        learning_rate=0.1,
        random_state=42
    )
}

# Evaluation Function
def evaluate_model(model, X_train, y_train, X_test, y_test):
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    # Training
    pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred = pipeline.predict(X_test)
    
    # Metrics
    metrics = {
        'MAE': mean_absolute_error(y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'R2': r2_score(y_test, y_pred)
    }
    
    return pipeline, metrics

# Model Comparison
results = {}
for name, model in tqdm(models.items()):
    pipeline, metrics = evaluate_model(model, X_train, y_train, X_test, y_test)
    results[name] = {
        'model': pipeline,
        'metrics': metrics
    }
    print(f"\n{name} Performance:")
    print(f"- MAE: {metrics['MAE']:,.2f}")
    print(f"- RMSE: {metrics['RMSE']:,.2f}")
    print(f"- R2: {metrics['R2']:.3f}")
```

## 5. Feature Importance Analysis

```python
# Get feature names from preprocessing
cat_encoder = preprocessor.named_transformers_['cat']
cat_features = cat_encoder.get_feature_names_out(categorical_features)
all_features = numeric_features + list(cat_features)

# Random Forest Feature Importance
rf_model = results['Random Forest']['model'].named_steps['model']
importances = pd.DataFrame({
    'Feature': all_features,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

# Visualization
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=importances.head(15))
plt.title("Top 15 Important Features (Random Forest)")
plt.tight_layout()
plt.savefig('../reports/figures/rf_feature_importance.png')
plt.show()
```

## 6. SHAP Value Analysis

```python
# Prepare SHAP data
best_model = results['XGBoost']['model']
X_processed = preprocessor.transform(X_test)

# Convert to dense array if sparse (THIS NEEDS TO COME FIRST)
if hasattr(X_processed, "toarray"):
    X_processed_dense = X_processed.toarray()
else:
    X_processed_dense = X_processed

# Create explainer and calculate SHAP values
explainer = shap.Explainer(best_model.named_steps['model'])
shap_values = explainer(X_processed_dense)  # Use the dense array here

# Set up random number generator
rng = np.random.default_rng(seed=42)

# Summary Plot
plt.figure()
shap.summary_plot(
    shap_values.values, 
    X_processed_dense,  # Now this variable exists
    feature_names=all_features,
    rng=rng
)
plt.title("SHAP Value Impact on Predictions")
plt.tight_layout()
plt.savefig('../reports/figures/shap_summary.png')
plt.show()

# Dependence Plot (Top Feature)
shap.dependence_plot(
    all_features[0],  # Use feature name directly
    shap_values.values, 
    X_processed_dense, 
    feature_names=all_features,
    interaction_index=None,
)
```

```python
# XGBoost Parameter Grid
param_grid = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [3, 6],
    'model__learning_rate': [0.01, 0.1]
}

# Create pipeline for tuning
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', XGBRegressor(random_state=42))
])

# Grid Search
scorer = make_scorer(mean_absolute_error, greater_is_better=False)
grid_search = GridSearchCV(
    pipe, 
    param_grid, 
    cv=3, 
    scoring=scorer,
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

# Best model results
print(f"Best MAE: {-grid_search.best_score_:,.2f}")
print("Best Parameters:", grid_search.best_params_)
```

```python
# Risk Segmentation Analysis
X_test['predicted_claims'] = best_model.predict(X_test)
X_test['actual_claims'] = y_test.values

# High vs Low Risk Groups
X_test['risk_group'] = pd.qcut(
    X_test['predicted_claims'],
    q=3,
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

# Premium Adjustment Simulation
X_test['recommended_premium'] = X_test['predicted_claims'] * 1.2  # 20% margin

# Save predictions
X_test[['predicted_claims', 'actual_claims', 'risk_group', 'recommended_premium']].to_csv(
    '../data/outputs/predictions.csv'
)

print("Prediction Sample:")
display(X_test[['predicted_claims', 'actual_claims']].describe())
```
