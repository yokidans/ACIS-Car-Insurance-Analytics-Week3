
# A/B Hypothesis Testing for Insurance Risk Analysis

This notebook focuses on conducting statistical hypothesis tests to evaluate:
- Differences in loss ratios across key segments (provinces, gender, zip codes)
- Statistical significance of observed variations
- Business implications of test results


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from tqdm import tqdm
import warnings
import os

# Configuration
warnings.filterwarnings('ignore')
pd.set_option('display.float_format', lambda x: '%.2f' % x)
sns.set_style("whitegrid")
sns.set_palette("husl")
```

```python
# Load the dataset with correct path
def load_data(file_path):
    """Function to safely load the dataset with error handling."""
    try:
        df = pd.read_csv(file_path, sep='|', encoding='ISO-8859-1')
        print("‚úÖ Data loaded successfully")
        return df
    except FileNotFoundError:
        print("‚ùå Error: File not found. Please check the file path.")
        return None
    except pd.errors.ParserError:
        print("‚ùå Error: Failed to parse the file. Ensure it's in CSV format.")
        return None
    except Exception as e:
        print(f"‚ùå Unexpected Error: {str(e)}")
        return None

# Load the dataset - REMOVED THE STRAY QUOTATION MARK
df = load_data('./data/raw/insurance_claims.csv')

# Exit if data wasn't loaded
if df is None:
    exit()

# Check basic info
print("\n=== DATA STRUCTURE ===")
print(df.info())

# Calculate Loss Ratio if needed
if 'TotalPremium' in df.columns and 'TotalClaims' in df.columns:
    df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
else:
    print("‚ùå Missing required columns for analysis")
    exit()

# Show basic info
print("\nüìã Data Overview:")
print(f"Columns: {df.columns.tolist()}")
print(f"Records: {len(df):,}")
print("\nMissing Values:")
print(df.isnull().sum())
```

## 1. Statistical Test Selection Guide

We'll use different tests based on data characteristics:

| Scenario                                  | Test Type               | Conditions                              | Implementation              |
|-------------------------------------------|-------------------------|-----------------------------------------|-----------------------------|
| Compare means across >2 groups            | ANOVA                   | Normal distribution, equal variances    | `scipy.stats.f_oneway`      |
| Compare means across 2 groups             | Independent t-test      | Normal distribution                     | `scipy.stats.ttest_ind`     |
| Compare medians across >2 groups          | Kruskal-Wallis          | Non-normal distributions                | `scipy.stats.kruskal`      |
| Compare proportions across categories     | Chi-square test         | Categorical data                        | `scipy.stats.chi2_contingency` |
| Compare paired samples                    | Paired t-test           | Before/after measurements               | `scipy.stats.ttest_rel`     |

## 2. Province Comparison (ANOVA)
    - Question: Do loss ratios differ significantly across provinces?
    -  Null Hypothesis: All provinces have equal loss ratios
    -  Alternative Hypothesis: At least one province differs
    -  Test: One-way ANOVA (parametric)
    -  Assumptions:
        - Normally distributed within groups
        - Homogeneity of variances

```python
def perform_ab_testing(df):
    """Conduct A/B hypothesis testing across key segments"""
    print("\nüî¨ Performing A/B Hypothesis Testing...")
    
    # Initialize results dictionary
    results = {}
    
    # Select top 5 provinces by policy count
    provinces = df['Province'].value_counts().nlargest(5).index
    groups = [df[df['Province'] == prov]['LossRatio'].dropna() for prov in provinces]
    
    print("\nüèôÔ∏è Testing provinces:", list(provinces))
    
    # Test normality (Shapiro-Wilk) and equal variance (Levene's)
    print("\nüìù Checking ANOVA assumptions:")
    normal = all(stats.shapiro(group)[1] > 0.05 for group in groups)
    equal_var = stats.levene(*groups)[1] > 0.05
    
    print(f"- Normally distributed: {'Yes' if normal else 'No'}")
    print(f"- Equal variances: {'Yes' if equal_var else 'No'}")
    
    # Perform ANOVA
    try:
        f_stat, p_value = stats.f_oneway(*groups)
        # Handle case where all groups are identical
        if np.isnan(f_stat):
            raise ValueError("All groups have identical means")
    except Exception as e:
        print(f"‚ö†Ô∏è ANOVA failed: {str(e)} - Using Kruskal-Wallis instead")
        # Fallback to non-parametric test
        f_stat, p_value = stats.kruskal(*groups, nan_policy='omit')
        test_used = 'Kruskal-Wallis'
    else:
        test_used = 'ANOVA'
    
    results['province_test'] = {
        'test': test_used,
        'comparison': 'Loss ratio across provinces',
        'groups': list(provinces),
        'statistic': round(f_stat, 2),
        'p_value': round(p_value, 4),
        'alpha': 0.05,
        'conclusion': "Reject null - significant difference" if p_value < 0.05 
                     else "Fail to reject - no significant difference",
        'assumptions_met': normal and equal_var if test_used == 'ANOVA' else 'N/A'
    }
    
    # Visualization
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Province', y='LossRatio', data=df[df['Province'].isin(provinces)])
    plt.title(f'Loss Ratio Distribution by Province\n({test_used} p-value: {p_value:.4f})')
    plt.axhline(y=df['LossRatio'].median(), color='r', linestyle='--', label='Overall Median')
    plt.legend()
    plt.show()
    
    # Create and display results table
    test_data = {
        "Test Type": [results['province_test']['test']],
        "Comparison": [results['province_test']['comparison']],
        "Groups": [", ".join(results['province_test']['groups'])],
        "Statistic": [results['province_test']['statistic']],
        "P-value": [results['province_test']['p_value']],
        "Significance": ["Significant" if results['province_test']['p_value'] < 0.05 
                        else "Not Significant"],
        "Assumptions Met": ["Yes" if results['province_test']['assumptions_met'] 
                           else "No" if test_used == 'ANOVA' else "N/A"]
    }
    
    results_table = pd.DataFrame(test_data)
    
    print("\nüìä Final Test Results Summary:")
    display(results_table.style
           .set_properties(**{'text-align': 'left'})
           .format({
               'Statistic': '{:.2f}',
               'P-value': '{:.4f}'
           }, na_rep="-")
           .hide(axis='index')
           .set_caption("Hypothesis Testing Results Summary"))
    
    return results

# Now run the function
ab_results = perform_ab_testing(df)
```

 ## 3. Gender Comparison (T-test)
    - **Question**: Is there a significant difference in loss ratios between genders?
    - **Null Hypothesis**: No difference in loss ratios (Œº_male = Œº_female)
    - **Alternative Hypothesis**: Loss ratios differ (Œº_male ‚â† Œº_female)
    - **Test**: Independent t-test (Welch's version for unequal variances)

```python
def perform_gender_comparison(df):
    """Conduct gender comparison analysis on loss ratios"""
    print("\nüöª Gender Comparison Analysis")
    
    # Initialize results dictionary
    results = {}
    
    # Check gender distribution
    gender_counts = df['Gender'].value_counts()
    print("\nGender distribution:")
    print(gender_counts)
    
    # Check if we have sufficient data for both genders
    valid_genders = [g for g in ['Male', 'Female'] if g in gender_counts.index and gender_counts[g] > 30]
    
    if len(valid_genders) < 2:
        print("‚ö†Ô∏è Insufficient gender groups for comparison (need at least Male and Female with >30 samples each)")
        return results
    
    # Prepare data - drop missing values
    male_loss = df[df['Gender'] == 'Male']['LossRatio'].dropna()
    female_loss = df[df['Gender'] == 'Female']['LossRatio'].dropna()
    
    # Check normality
    print("\nüìù Normality checks:")
    male_normal = stats.shapiro(male_loss)[1] > 0.05
    female_normal = stats.shapiro(female_loss)[1] > 0.05
    print(f"- Male normality: {'Yes' if male_normal else 'No'} (p={stats.shapiro(male_loss)[1]:.4f})")
    print(f"- Female normality: {'Yes' if female_normal else 'No'} (p={stats.shapiro(female_loss)[1]:.4f})")
    
    # Check variance equality
    _, p_var = stats.levene(male_loss, female_loss)
    use_equal_var = p_var > 0.05
    print(f"\nüìù Variance equality: {'Equal' if use_equal_var else 'Unequal'} (Levene's p={p_var:.4f})")
    
    # Select appropriate test
    if male_normal and female_normal:
        # Parametric t-test
        t_stat, p_value = stats.ttest_ind(male_loss, female_loss, equal_var=use_equal_var)
        test_used = "Independent t-test"
    else:
        # Non-parametric alternative
        t_stat, p_value = stats.mannwhitneyu(male_loss, female_loss)
        test_used = "Mann-Whitney U test"
        print("‚ö†Ô∏è Using non-parametric test due to non-normal distributions")
    
    # Calculate effect size (Cohen's d)
    pooled_std = np.sqrt((male_loss.std()**2 + female_loss.std()**2)/2)
    cohens_d = (male_loss.mean() - female_loss.mean()) / pooled_std
    
    # Store results
    results['gender_test'] = {
        'test': test_used,
        'comparison': 'Loss ratio by gender',
        'groups': valid_genders,
        'statistic': round(t_stat, 2),
        'p_value': round(p_value, 4),
        'alpha': 0.05,
        'effect_size': round(cohens_d, 2),
        'conclusion': "Reject null - significant difference" if p_value < 0.05 
                     else "Fail to reject - no significant difference",
        'assumptions_met': {
            'normality': male_normal and female_normal,
            'equal_variance': use_equal_var
        }
    }
    
    # Visualization
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Gender', y='LossRatio', data=df[df['Gender'].isin(valid_genders)])
    plt.title(f'Loss Ratio by Gender\n({test_used} p-value: {p_value:.4f})')
    plt.axhline(y=df['LossRatio'].median(), color='r', linestyle='--', label='Overall Median')
    plt.legend()
    plt.show()
    
    return results

# Now run the function
gender_results = perform_gender_comparison(df)

# Display results
print("\nüìä Gender Test Results Summary:")
pd.DataFrame(gender_results['gender_test'].items(), columns=['Metric', 'Value']).style.hide(axis='index')
```

 ## 4. Zip Code Comparison (Kruskal-Wallis)
    - **Question**: Do profit margins differ across high-volume zip codes?
    - **Null Hypothesis**: All zip codes have equal profit margins
    - **Alternative Hypothesis**: At least one zip code differs
    - **Test**: Kruskal-Wallis (non-parametric alternative to ANOVA)

```python
# Initialize results dictionary to store test outputs
results = {}

# Filter for zip codes with sufficient data
MIN_POLICIES = 50
zip_counts = df['PostalCode'].value_counts()
valid_zips = zip_counts[zip_counts >= MIN_POLICIES].index

# Check if we have ProfitMargin or should use LossRatio
if 'ProfitMargin' not in df.columns:
    if 'TotalPremium' in df.columns and 'TotalClaims' in df.columns:
        # Calculate ProfitMargin if we have the components
        df['ProfitMargin'] = (df['TotalPremium'] - df['TotalClaims']) / df['TotalPremium']
        metric = 'ProfitMargin'
    else:
        # Fall back to LossRatio if we can't calculate ProfitMargin
        if 'LossRatio' in df.columns:
            metric = 'LossRatio'
        else:
            print("‚ùå No suitable metric available for comparison (need ProfitMargin or LossRatio)")
            exit()
else:
    metric = 'ProfitMargin'

if len(valid_zips) < 2:
    print(f"‚ö†Ô∏è Only {len(valid_zips)} zip codes with ‚â•{MIN_POLICIES} policies")
else:
    top_zips = valid_zips[:5]  # Limit to top 5 for clarity
    groups = [df[df['PostalCode'] == zipcode][metric] for zipcode in top_zips]
    
    h_stat, p_value = stats.kruskal(*groups)
    
    results['zipcode_kruskal'] = {
        'test': 'Kruskal-Wallis',
        'comparison': f'{metric} by zip code',
        'groups': list(top_zips),
        'h_stat': round(h_stat, 2),
        'p_value': round(p_value, 4),
        'alpha': 0.05,
        'conclusion': "Reject null - significant difference" if p_value < 0.05 
                     else "Fail to reject - no significant difference"
    }
    
    # Visualization
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='PostalCode', y=metric, 
              data=df[df['PostalCode'].isin(top_zips)])
    plt.title(f'{metric} Distribution by Zip Code\n(Kruskal-Wallis p-value: {p_value:.4f})')
    plt.xticks(rotation=45)
    plt.show()
```

## 5. Vehicle Type vs Claim Frequency (Chi-square)
    - **Question**: Is there an association between vehicle type and claim occurrence?
    - **Null Hypothesis**: No association between vehicle type and claim frequency
    - **Alternative Hypothesis**: Association exists
    - **Test**: Chi-square test of independence

```python
    # Create binary claim flag and contingency table
    df['HadClaim'] = df['TotalClaims'] > 0
    contingency = pd.crosstab(df['VehicleType'], df['HadClaim'])
    
    # Filter for vehicle types with sufficient data
    contingency = contingency[contingency.sum(axis=1) >= 50]  # Min 50 policies
    
    if len(contingency) < 2:
        print("‚ö†Ô∏è Insufficient vehicle types for comparison")
    else:
        chi2, p_value, _, _ = stats.chi2_contingency(contingency)
        
        # Calculate Cramer's V effect size
        n = contingency.sum().sum()
        phi = np.sqrt(chi2/n)
        min_dim = min(contingency.shape)-1
        cramers_v = phi / np.sqrt(min_dim)
        
        results['vehicletype_chi2'] = {
            'test': 'Chi-square test',
            'comparison': 'Claim frequency by vehicle type',
            'chi2_stat': round(chi2, 2),
            'p_value': round(p_value, 4),
            'alpha': 0.05,
            'effect_size': round(cramers_v, 2),
            'conclusion': "Reject null - significant association" if p_value < 0.05 
                         else "Fail to reject - no significant association"
        }
        
        # Visualization
        plt.figure(figsize=(12, 6))
        sns.heatmap(contingency.div(contingency.sum(1), axis=0), 
                   annot=True, fmt='.1%', cmap='Blues')
        plt.title('Claim Frequency % by Vehicle Type\n(œá¬≤ p-value: {:.4f})'.format(p_value))
        plt.show()
```

```python
def generate_ab_recommendations(test_results, df):
    """Generate business recommendations from test results"""
    print("\nüí° Business Recommendations:")
    recommendations_made = False
    
    # Province analysis
    prov_result = test_results.get('province_anova')
    if prov_result and 'p_value' in prov_result:
        if prov_result['p_value'] < 0.05:
            print("\nüîπ PROVINCE IMPACT: SIGNIFICANT (p={:.4f})".format(prov_result['p_value']))
            if 'Province' in df.columns and 'LossRatio' in df.columns:
                print("   - Top 3 provinces by loss ratio:")
                province_stats = df.groupby('Province')['LossRatio'].median().sort_values(ascending=False)
                print(province_stats.head(3).to_string())
            print("\n   ACTIONS:")
            print("   ‚Ä¢ Implement province-specific pricing strategies")
            print("   ‚Ä¢ Allocate more underwriting resources to high-risk provinces")
            print("   ‚Ä¢ Investigate regional risk factors (weather, theft rates)")
            recommendations_made = True
        else:
            print("\nüîπ PROVINCE IMPACT: NOT SIGNIFICANT")
            print("   - Consider standardizing pricing across provinces")
            recommendations_made = True

    # Gender analysis
    gender_result = test_results.get('gender_ttest')
    if gender_result and 'p_value' in gender_result:
        if gender_result['p_value'] < 0.05:
            print("\nüîπ GENDER IMPACT: SIGNIFICANT (p={:.4f})".format(gender_result['p_value']))
            if 'effect_size' in gender_result:
                print(f"   - Effect size (Cohen's d): {gender_result['effect_size']:.2f}")
            if 'Gender' in df.columns and 'LossRatio' in df.columns:
                gender_stats = df.groupby('Gender')['LossRatio'].agg(['median', 'count'])
                print(gender_stats.to_string())
            print("\n   ACTIONS:")
            print("   ‚Ä¢ Evaluate gender-based pricing if legally permissible")
            print("   ‚Ä¢ Develop targeted marketing campaigns")
            print("   ‚Ä¢ Investigate behavioral differences in claims")
            recommendations_made = True
        else:
            print("\nüîπ GENDER IMPACT: NOT SIGNIFICANT")
            print("   - Gender-neutral pricing is statistically justified")
            recommendations_made = True

    # Zip code analysis
    zip_result = test_results.get('zipcode_kruskal')
    if zip_result and 'p_value' in zip_result:
        if zip_result['p_value'] < 0.05:
            print("\nüîπ ZIP CODE IMPACT: SIGNIFICANT (p={:.4f})".format(zip_result['p_value']))
            if 'PostalCode' in df.columns and 'ProfitMargin' in df.columns:
                print("   - Top/Bottom 3 zip codes by profit margin:")
                zip_stats = df.groupby('PostalCode')['ProfitMargin'].median().sort_values()
                print(pd.concat([zip_stats.head(3), zip_stats.tail(3)]).to_string())
            print("\n   ACTIONS:")
            print("   ‚Ä¢ Implement granular geographic pricing")
            print("   ‚Ä¢ Investigate local risk factors in high-claim zip codes")
            print("   ‚Ä¢ Partner with local repair shops in high-cost areas")
            recommendations_made = True

    # Vehicle type analysis
    vehicle_result = test_results.get('vehicletype_chi2')
    if vehicle_result and 'p_value' in vehicle_result:
        if vehicle_result['p_value'] < 0.05:
            print("\nüîπ VEHICLE TYPE IMPACT: SIGNIFICANT (p={:.4f})".format(vehicle_result['p_value']))
            if 'effect_size' in vehicle_result:
                print(f"   - Effect size (Cramer's V): {vehicle_result['effect_size']:.2f}")
            if 'VehicleType' in df.columns and 'HadClaim' in df.columns:
                print("   Claim frequency by vehicle type:")
                vehicle_stats = df.groupby('VehicleType')['HadClaim'].mean().sort_values(ascending=False)
                print(vehicle_stats.to_string())
            print("\n   ACTIONS:")
            print("   ‚Ä¢ Adjust premiums by vehicle type based on risk")
            print("   ‚Ä¢ Offer discounts for low-risk vehicle types")
            print("   ‚Ä¢ Partner with manufacturers of safe vehicles")
            recommendations_made = True

    if not recommendations_made:
        print("\nNo actionable recommendations could be generated from the test results.")
        print("Possible reasons:")
        print("- No significant differences found in any tests")
        print("- Required data columns missing from the dataset")
        print("- Test results not in expected format")

# Example usage:
#generate_ab_recommendations(results, df)
```
