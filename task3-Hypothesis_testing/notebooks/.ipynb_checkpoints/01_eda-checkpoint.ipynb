# Exploratory Data Analysis for ACIS Car Insurance Analytics
## Comprehensive Data Exploration and Visualization

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from IPython.display import display

# Set visualization style
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
np.random.seed(42)
```

```python
# Load preprocessed data
df = pd.read_csv('../data/processed/processed_insurance_data.csv')
```

## 1. Data Overview

```python
print("Dataset Shape:", df.shape)
display(df.head(3))

# Data types and missing values
print("\nData Types and Missing Values:")
display(pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': df.isnull().sum(),
    'Unique Values': df.nunique()
}))

# Descriptive statistics
print("\nDescriptive Statistics:")
display(df.describe(include=[np.number]).T.style\
       .background_gradient(cmap='Blues', subset=['mean', '50%', 'std']))
```

## 2. Premium Analysis

```python
## 2. Premium Analysis
plt.figure(figsize=(15, 5))

# Premium distribution with log transform
plt.subplot(1, 2, 1)
# Filter out negative premiums before log transform
valid_premiums = df['totalpremium'][df['totalpremium'] > 0]
if len(valid_premiums) > 0:
    log_premiums = np.log1p(valid_premiums)
    sns.histplot(log_premiums, bins=50, kde=True)
    plt.title('Log-Transformed Premium Distribution (Positive Premiums Only)')
    plt.xlabel('Log(Premium + 1)')
else:
    plt.text(0.5, 0.5, 'No valid positive premiums to plot', 
             ha='center', va='center')
    plt.title('No Data Available')

# Premium by vehicle type (top 10)
plt.subplot(1, 2, 2)
top_vehicle_types = df['vehicletype'].value_counts().nlargest(10).index
sns.boxplot(
    x='vehicletype', 
    y='totalpremium', 
    data=df[df['vehicletype'].isin(top_vehicle_types)],
    order=top_vehicle_types
)
plt.xticks(rotation=45)
plt.title('Premium Distribution by Vehicle Type (Top 10)')
plt.tight_layout()
plt.show()

# Report any invalid premiums
negative_premiums = (df['totalpremium'] <= 0).sum()
if negative_premiums > 0:
    print(f"⚠️ Warning: Found {negative_premiums} policies with non-positive premiums")
    print("These have been excluded from the log-transformed distribution plot")
```

## 3. Loss Ratio Deep Dive

```python
plt.figure(figsize=(15, 5))

# Loss ratio distribution
plt.subplot(1, 2, 1)
sns.histplot(df['loss_ratio'], bins=50, kde=True)
plt.axvline(x=1, color='r', linestyle='--')
plt.title('Loss Ratio Distribution (1 = Break-even)')

# Loss ratio by province
plt.subplot(1, 2, 2)
province_loss = df.groupby('province', observed=True)['loss_ratio'].mean().sort_values()
top_bottom = pd.concat([province_loss.head(5), province_loss.tail(5)])
sns.barplot(x=top_bottom.values, 
            y=top_bottom.index, 
            hue=top_bottom.index,  # Added hue parameter
            palette='RdYlGn',
            legend=False)          # Disabled legend
plt.title('Top 5 (Green) & Bottom 5 (Red) Provinces by Loss Ratio')
plt.tight_layout()
plt.show()
```

## 4. Vehicle Age Analysis

```python
## 4. Vehicle Age Analysis
# Create age groups
df['age_group'] = pd.cut(df['vehicle_age'], 
                        bins=[0, 3, 5, 7, 10, 15, 30],
                        labels=['0-3', '4-5', '6-7', '8-10', '11-15', '15+'])

plt.figure(figsize=(15, 5))

# Claims frequency by age group
plt.subplot(1, 2, 1)
age_claims = df.groupby('age_group', observed=True)['totalclaims'].mean()
sns.barplot(x=age_claims.index, y=age_claims.values)
plt.title('Average Claims by Vehicle Age Group')
plt.ylabel('Average Claims Amount')

# Loss ratio by age group
plt.subplot(1, 2, 2)
age_loss = df.groupby('age_group', observed=True)['loss_ratio'].mean()
sns.barplot(x=age_loss.index, y=age_loss.values)
plt.title('Loss Ratio by Vehicle Age Group')
plt.tight_layout()
plt.show()
```

## 5. Correlation Analysis

```python
# Select numerical features
num_features = ['totalpremium', 'totalclaims', 'suminsured', 
               'customvalueestimate', 'vehicle_age', 'loss_ratio']
corr_matrix = df[num_features].corr()

# Visualize with heatmap
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', 
           center=0, fmt='.2f', linewidths=.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

## 7. Categorical Feature Analysis

```python
# Vehicle type profitability
vehicle_profit = df.groupby('vehicletype').agg({
    'totalpremium': 'mean',
    'totalclaims': 'mean',
    'loss_ratio': 'mean',
    'policyid': 'count'
}).sort_values('loss_ratio', ascending=False)

vehicle_profit['profit_margin'] = vehicle_profit['totalpremium'] - vehicle_profit['totalclaims']
vehicle_profit = vehicle_profit.rename(columns={'policyid': 'policy_count'})

# Display top/bottom 5
print("Vehicle Type Profitability Analysis:")
display(vehicle_profit.head(5).style.background_gradient(cmap='Reds', subset=['loss_ratio']))
display(vehicle_profit.tail(5).style.background_gradient(cmap='Greens', subset=['loss_ratio']))
```

## 8. Geographic Analysis

```python
if 'province' in df.columns:
    geo_analysis = df.groupby('province').agg({
        'totalpremium': ['mean', 'sum'],
        'totalclaims': ['mean', 'sum'],
        'loss_ratio': 'mean',
        'policyid': 'count'
    }).sort_values(('loss_ratio', 'mean'), ascending=False)
    
    geo_analysis.columns = ['_'.join(col).strip() for col in geo_analysis.columns.values]
    geo_analysis = geo_analysis.rename(columns={'policyid_count': 'policy_count'})
    
    print("\nGeographic Profitability Analysis:")
    display(geo_analysis.head(5).style.background_gradient(cmap='Reds', subset=['loss_ratio_mean']))
    display(geo_analysis.tail(5).style.background_gradient(cmap='Greens', subset=['loss_ratio_mean']))
```

## 9. Outlier Detection

```python
# Identify high-loss ratio policies
high_loss = df[df['loss_ratio'] > 2]  # Policies where claims are 2x premiums
if not high_loss.empty:
    print(f"\nFound {len(high_loss)} extreme loss policies (loss ratio > 2):")
    display(high_loss.describe(include=[np.number]).T)
    
    # Analyze characteristics
    print("\nCharacteristics of High-Loss Policies:")
    for col in ['vehicletype', 'province', 'age_group']:
        if col in high_loss.columns:
            print(f"\n{col} distribution in high-loss policies:")
            display(high_loss[col].value_counts(normalize=True).head(5))
```

# Key Insights Summary

1. **Premium Distribution**: Right-skewed with significant variation across vehicle types
2. **Loss Ratios**: X% of policies are unprofitable (loss ratio > 1)
3. **Vehicle Age**: U-shaped risk pattern with highest risk at newest and oldest vehicles
4. **Geographic Patterns**: Clear regional differences in profitability
5. **Vehicle Types**: Identified both high-risk and consistently profitable categories
6. **Correlations**: Premiums correlate strongly with sum insured but weakly with actual claims
